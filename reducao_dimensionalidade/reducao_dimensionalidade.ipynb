```python
dataset_dir = '/content/imagens'
all_files = os.listdir(dataset_dir)

gray_list = []

def convert_gray(image):
  for i in (all_files):
    image_path = os.path.join(dataset_dir, i)
    image_pil = Image.open(image_path)
    image_array = np.array(image_pil)
    r, g, b = image_array[:,:, 0], image_array[:,:, 1], image_array[:,:, 2]
    gray = ((0.3 * r) + (0.6 * g) + (0.1 * b))
    gray = gray.astype(np.uint8)

    gray_list.append(gray)
  return gray_list
```

```python
converted_image = convert_gray(all_files)
for gray_img in converted_image:
    plt.imshow(gray_img, cmap='gray')
    plt.show()
```

```python
binary_list = []

def binary_color(image, threshold=128):
  for i in gray_list:
    image = i
    binary_image = np.where(image >= threshold, 255, 0)
    binary_image = binary_image.astype(np.uint8)
    binary_list.append(binary_image)
  return binary_list
```

```python
converted_image = binary_color(gray_list)
for binary_img in converted_image:
    plt.imshow(binary_img, cmap='gray')
    plt.show()
```
